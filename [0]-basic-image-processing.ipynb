## Creating Shades of Red - Dence

```python
img = []
for i in range(1080) :
    t = []
    for j in range(1080) :   # Black to redd formation
        t.append([i,0,0])
    img.append(t)
for i in range(255,0,-1) :
    t = []
    for j in range(1080) :    # Redd to Black formation
        t.append([i,0,0]) 
    img.append(t)

for i in range(1080) :
    t = []
    for j in range(1080) :   # RBKS to redd formation
        t.append([0,i,0])
    img.append(t)
for i in range(0,i,-1) :
    t = []
    for j in range(1080) :    # Redd to Black formation
        t.append([0,i,0]) 
    img.append(t)
     
    
iimg = np.array(img)
plt.imshow(iimg)
```
